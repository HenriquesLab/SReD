```python
## ------------------------------------------------------------------ ##
## ---- Plot 3D spot detections with depth color-coded (example) ---- ##
## ------------------------------------------------------------------ ##

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV into a pandas DataFrame
df = pd.read_csv('/Volumes/OCB-AM/Image_library/2023/20231222_jurkat_tet-iGFP_puro_cleanse/cell1/beads/input_spots.csv')

# Extract X, Y and Z coordinates
x = df['X']
y = df['Y']
z = df['Z']

# Plot the points
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)

# Define the colormap
cmap = plt.cm.jet

# Plot the scatter plot with z-coordinate mapped to color
scatter = ax.scatter(x, y, c=z, cmap='viridis_r', vmin=3, vmax=41) # Adjust vmax and vmin

# Add color bar
#cbar = plt.colorbar(scatter, ax=ax)
#cbar.set_label('Z', fontsize=16)

# Add labels
ax.set_xlabel('X', fontsize=16)
ax.set_ylabel('Y', fontsize=16)

# Adjust plot limits
ax.set_xlim(0, 314)
ax.set_ylim(0, 314)

plt.savefig('/Volumes/OCB-AM/Image_library/2023/20231222_jurkat_tet-iGFP_puro_cleanse/cell1/beads/input_spots.png', dpi=300)
```
