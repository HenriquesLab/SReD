```python
## ------------------------------ ##
## ---- Polar plot (example) ---- ##
## ------------------------------ ##

# Import libraries
import matplotlib.pyplot as plt
import numpy as np

# Define data (average SRS for each of the eight orientations, starting on top and going clockwise)
values = [0.36, 0.25, 0.33, 0.26, 0.3, 0.24, 0.35, 0.2]

# Number of segments
num_segments = 8

# Calculate the width of each segment
segment_width = 360 / num_segments

# Rotation angle in degrees
rotation_angle_degrees = 113  # Change this value as needed, it just changes the plot's alignment

# Convert rotation angle to radians
rotation_angle_radians = np.radians(rotation_angle_degrees)

# Initialize the figure and axis
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})

# Initialize the angle
angle = 0

# Define colors for each segment
colors = ['magenta', 'gold', 'red', 'gold', 'steelblue', 'gold', 'limegreen', 'gold']

# Plot each segment
for value, color in zip(values, colors):
    # Convert angle to radians and add the rotation angle
    theta = np.radians(angle) + rotation_angle_radians
    
    # Calculate the end angle
    end_theta = np.radians(angle - segment_width) + rotation_angle_radians
    
    # Plot the segment with specific color
    ax.fill_between([theta, end_theta], 0, value, alpha=1, color=color)
    
    # Decrease the angle for the next segment
    angle -= segment_width

# Set the title
ax.set_title('Rose Diagram')

# Adjust the grid
ax.set_thetagrids(np.arange(0, 360, 45), [])
ax.set_rgrids(np.arange(0, 0.25, 1), [])

# Hide all axis labels
ax.set_xticklabels([])
ax.set_yticklabels([])

# Set radius limit
#plt.ylim(0, 1.0)

# Show the plot
plt.show()
#plt.savefig('*/round.png', dpi=300) # Change directory to path of interest


```
